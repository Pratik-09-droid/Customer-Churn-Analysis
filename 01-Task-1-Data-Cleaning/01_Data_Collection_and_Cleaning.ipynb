```python
df[df["TotalCharges"]==" "]    #So this line shows all the 11 rows in which the 'TotalCharges' has blank value
```

```python
#Here you can see that all have tenure 0  which means
#the customer just joined the company they have't completed one month
df[df["TotalCharges"]==" "][["customerID" , "tenure" ,"MonthlyCharges","TotalCharges","Churn" ]]
```

```python
# Now from this line we replace the blank space with 'nan' value
df["TotalCharges"] = df["TotalCharges"].replace(" " , np.nan)
```

If an interviewer asks:

"How did you identify missing values in TotalCharges?"

A professional answer is:

"Although df.info() reported no null values, I noticed that TotalCharges was stored as an object instead of a numeric type. On further inspection, I found that 11 records contained a single space (' ') rather than actual null values. I verified this using df[df['TotalCharges'] == ' '] and then handled these values during data cleaning."

 Now we convert the'TotalChrges' dtype from object to numeric as this column store the Total charges which will be a number so if it will a object we can not do a statistics such finding mean , avg etc so we have to convert it into a float datatype

 in second line we are feeling **0** in place of **nan** ,the fillna function only fill value in a nan part that's why we first replace blank space with nan

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"])
df["TotalCharges"] = df["TotalCharges"].fillna(0)
```

How did you handle missing values in TotalCharges?

You can confidently say:

"The TotalCharges column contained 11 blank strings (' ') instead of actual null values, so Pandas treated the column as an object. I first converted the blank strings to NaN, inspected the affected rows, and found that all of them had tenure = 0, meaning they were new customers who had not yet completed a billing cycle. Based on this business logic, I replaced the missing TotalCharges with 0 and converted the column to a numeric type."
